```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

# Генерация

```python
sizes = [100, 500, 1000, 5000, 10000, 50000, 100000]
names = []
midnames = []
surnames = []
departments = [f'Отдел{i}' for i in range(1, 13)]
positions = ['Начальник', 'Секретарь', 'Менеджер', 'Старший сотрудник', 'Сотрудник', 'Младший сотрудник']
```

```python
f = open('names_table.txt', 'r')
for line in f:
    names.append(line.split('"')[3])
#-----------------------------------------
f = open('midnames_table.txt', 'r')
for line in f:
    midnames.append(line.split('"')[3])
#-----------------------------------------
f = open('surnames_table.txt', 'r')
for line in f:
    surnames.append(line.split('"')[3])
```

```python
for size in sizes:
    f = open(f'dataset_{size}.txt', 'w')
    for i in range(size):
        surname = random.choice(surnames)
        name = random.choice(names)
        midname = random.choice(midnames)
        dep = random.choice(departments)
        pos = random.choice(positions)
        year = random.randint(2003, 2023)
        month = random.randint(1, 12)
        day = random.randint(1, 28)
        f.write(f'{surname} {name} {midname};{dep};{pos};{year}-{month}-{day}\n')
        if (i==size-1):
            f.write(f'{surname} {name} {midname};{dep};{pos};{year}-{month}-{day}')
    f.close()
```

# ЛР1

```python
bubble_diffs = np.array([0.000258963, 0.0277575, 0.697492, 2.77905, 8.42516, 73.5077, 334.133])
heap_diffs = np.array([7.1564e-05, 0.000937328, 0.00599418, 0.0133138, 0.00438206, 0.109025, 0.294229])
merge_diffs = np.array([0.000143248, 0.00199108, 0.0165444, 0.0300573, 0.0372886, 0.248564, 0.609219])

bubble_diffs_ln = np.log(bubble_diffs)
heap_diffs_ln = np.log(heap_diffs)
merge_diffs_ln = np.log(merge_diffs)
```

```python
plt.title('Зависимости времени сортировки от размерности массива')
plt.xlabel('Размерность массива')
plt.ylabel('Время, сек')
plt.plot(sizes, bubble_diffs, color='r', marker='o')
plt.plot(sizes, heap_diffs, color='g', marker='s')
plt.plot(sizes, merge_diffs, color='y', linestyle='--', marker='D')
plt.legend(['bubble sort', 'heap sort', 'merge sort'])
plt.show()
```

```python
plt.title('Зависимости времени сортировки от размерности массива')
plt.xlabel('Размерность массива')
plt.ylabel('Время, сек')
plt.plot(sizes, heap_diffs, color='g', marker='s')
plt.plot(sizes, merge_diffs, color='y', linestyle='--', marker='D')
plt.legend(['heap sort', 'merge sort'])
plt.show()
```

```python
plt.title('Зависимости времени сортировки от размерности массива')
plt.xlabel('Размерность массива')
plt.ylabel('ln от времени, ln(сек)')
plt.plot(sizes, bubble_diffs_ln, color='r', marker='o')
plt.plot(sizes, heap_diffs_ln, color='g', marker='s')
plt.plot(sizes, merge_diffs_ln, color='y', linestyle='--', marker='D')
plt.legend(['bubble sort', 'heap sort', 'merge sort'])
plt.show()
```
